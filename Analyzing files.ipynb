```python
import json

pythonListofDicts = []

for data in content1:
    pythonListofDicts.append(json.loads(data))
```

```python
pythonListofDicts[-1]
```

```python
pythonListofDicts[-1]['stream_type']
```

```python
bvp_value2 = []
bvp_timestamp2 = []

for i in range(len(pythonListofDicts)):
    sample = pythonListofDicts[i]
    if sample['stream_type'] == 'E4_Bvp':
        bvp_value2.append(sample['Value'])
        bvp_timestamp2.append(sample['dateTime'])

print(len(bvp_value2), len(bvp_timestamp2))
```

```python
bvp_value = []
bvp_timestamp = []

for i in range(len(pythonListofDicts)):
    if pythonListofDicts[i]['stream_type'] == 'E4_Bvp':
        bvp_value.append(float(pythonListofDicts[i]['Value']))
        bvp_timestamp.append(pythonListofDicts[i]['dateTime'])

print(len(bvp_value), len(bvp_timestamp))
```

```python
import pandas as pd

df = pd.DataFrame()

df['BVP_Value'] = bvp_value
df['DateTime'] = bvp_timestamp
```
